### File management


```
file = open(filename,option)
file.close()

option:
- r  : Open the file (if it exist) in reading mode (default)
- r+ : If you want to read and write to a file (if it exist).
- w  : To overwrite the content (create the file if it does not exist).
- a  : To add to the end of the file if it already exists.
- b  : Open the file in binary form
```

# Functions

### _code reuse principle_
```
def my_function():
    ...code

```
| Params | Output | Examples |
| -- | -- | -- |
| No | No | ``` def my_function():  ``` |
| No | Yes |``` def my_function(): return value  ``` |
| Yes | No |``` def my_function(param1, param2):  ``` |
| Yes | Yes |``` def my_function(param1, param2): return value ``` |


````python
# function without params and without output
def my_first_function():
    print("Hello from my first function in Python!!!")
````

````python
my_first_function()
````

````python
# function without params with output
def my_second_function():
    even_numbers = [number for number in range(100) if number % 2 == 0]
    return even_numbers
````

````python
numbers = my_second_function()
print(numbers)
````

````python
def even_and_odd():
    even_numbers = [number for number in range(100) if number % 2 == 0]
    odd_numbers = [number for number in range(100) if number % 2 != 0]
    return even_numbers, odd_numbers
````

# SOLID

````python
numbers = even_and_odd()
print(numbers)
````

````python
even, odd = even_and_odd()
print(even)
print(odd)
````

### Fressnel Zone
<img src="fresnelzone.png">
$ r_n = 17.32 * \sqrt{\left(\dfrac{n * d_1 * d_2}{f * (d_1 + d_2)}\right)} $

````python
import math
````

````python
def calculate_fresnel_zone(d1, d2, freq, n=1):
    rn = (n * d1 * d2) / (freq * (d1 + d2))
    rn = 17.32 * math.sqrt(rn)
    return rn
````

````python
rn = calculate_fresnel_zone(1000, 1000, 2437)
print(rn)
````

````python
rn = calculate_fresnel_zone(freq = 2437, d2 = 1000, d1 = 1000)
print(rn)
````

````python
calculate_fresnel_zone(d1 = 1000, d2 = 1000, 2437)
````

````python
calculate_fresnel_zone(1000,1000,freq = 2437)
````

## Function with infite params

````python
def infinite_by_position(*args):
    print(args)
    for arg in args:
        print(arg)
````

````python
infinite_by_position(1,2,3,4,5,6,7,8,9,0,1,3,56,7,8,5,4,34,7,7,5,4,4,6,1,1,1,1,11,11)
````

````python
def inifinite_by_keyword(**kwargs):
    print(kwargs)
    for kwarg in kwargs:
        print(kwarg, kwargs[kwarg])
````

````python
inifinite_by_keyword(number1=1, number2=2, number3=3, number4=4, number5=5)
````
